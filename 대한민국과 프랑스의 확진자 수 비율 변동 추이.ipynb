```python
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://covid.ourworldindata.org/data/owid-covid-data.csv'
df = pd.read_csv(url, parse_dates=['date'])

mask = (
    df['iso_code'].isin(['KOR', 'FRA']) &
    (df['date'] >= '2020-01-22') &
    (df['date'] <= '2022-04-30')
)
df2 = df.loc[mask, ['date', 'iso_code', 'total_cases']].copy()

grouped = df2.groupby('iso_code')

plt.figure(figsize=(10, 6))
for iso, grp in grouped:
    label = 'KOR' if iso == 'KOR' else 'FRA'
    plt.plot(grp['date'], grp['total_cases'], label=label)

plt.title('대한민국(KOR) vs 프랑스(FRA)의 일별 누적 확진자 수\n2020-01-22 ~ 2022-04-30')
plt.xlabel('date')
plt.ylabel('total cases')
plt.legend()
plt.tight_layout()
plt.show()
```
